```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifa2022=pd.read_csv("~/FIFA22_official_data.csv")
```

```python
fifa2022.query("Nationality=='India'").groupby(['Club']).sum().plot(kind='bar', y='Overall',colormap='Paired')
plt.xlabel("Team")
plt.ylabel("Overall Player Performance")
plt.title("Total Player Performance Per Team")
```
