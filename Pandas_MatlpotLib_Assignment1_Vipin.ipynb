```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifa2022=pd.read_csv("~/FIFA22_official_data.csv")
```

```python
fifa2022.query("Nationality=='India'").groupby(['Club']).sum().plot(kind='bar', y='Overall',color='green')
plt.xlabel("Team")
plt.ylabel("Overall Player Performance")
plt.title("Total Player Performance Per Team")
```

```python
fifa2022=fifa2022.assign(Player_Position=fifa2022["Position"].str.extract(r'>(\w+)$',expand = True))
```

```python
fifa2022.query("Nationality=='India' and Player_Position=='GK'").groupby(['Name']).sum().plot(kind='barh', y='GKReflexes',color='blue')
plt.xlabel("Team")
plt.ylabel("GK Reflexes")
plt.title("GK With Best Reflexes")
```
